**Задание 1**   
Постройте тензор ранга 1 (вектор) со значениями функции одной переменной $f(x) = (x^3+2) e^{-x}$ на отрезке [0,5] и определите максимальное и минимальное значения функции.

```python
import tensorflow as tf
import numpy as np

rank_1_tensor = tf.constant(np.linspace(0,5,num=100), dtype=tf.float32)

with tf.GradientTape() as tape:
  tape.watch(rank_1_tensor)
  y = (rank_1_tensor**3 + 2) * np.exp(-rank_1_tensor)

print("Значения функции f(x) на интервале [0, 5]:\n")
print(y)
# print(rank_1_tensor)
rank_1_tensor.ndim
```

```python
max = np.max(y)
min = np.min(y)

print("Максимальное значение ф-ции:", max)
print("Минимальное значение ф-ции:", min)
```

```python
import matplotlib.pyplot as plt

plt.plot(rank_1_tensor, y, label='f(x)')
plt.axhline(y=max, color='r', label='Максимальное значение')
plt.axhline(y=min, color='m', label='Минимальное значение')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x) с прямыми')
plt.legend()
plt.grid(True)
plt.show()
```

**Задание 3**   
Найдите значения производной от функции 4 порядка и постройте график полученной функции, подписывая оси и рисунок.

```python
with tf.GradientTape() as gt1:
    gt1.watch(rank_1_tensor)
    with tf.GradientTape() as gt2:
        gt2.watch(rank_1_tensor)
        with tf.GradientTape() as gt3:
            gt3.watch(rank_1_tensor)
            with tf.GradientTape() as gt4:
                gt4.watch(rank_1_tensor)
                yy = (rank_1_tensor**3+2)*tf.exp(-rank_1_tensor)
            dydx = gt4.gradient(yy, rank_1_tensor)
        d2ydx2 = gt3.gradient(dydx , rank_1_tensor)
    d3ydx3 = gt2.gradient(d2ydx2, rank_1_tensor)
d4ydx4 = gt1.gradient(d3ydx3 , rank_1_tensor)

# print('dy/dx =', dydx)
# print('d2y/dx2 =', d2ydx2)
# print('d3y/dx3 =', d3ydx3)
print('d4y/dx4 =', d4ydx4)
```

```python
with tf.GradientTape() as gt1:
    gt1.watch(obl2)
    with tf.GradientTape() as gt2:
        gt2.watch(obl1)
        with tf.GradientTape() as gt3:
            gt3.watch(obl1)
            with tf.GradientTape() as gt4:
                gt4.watch(obl1)
                yz = tf.exp(-obl1)*tf.sin(obl2)
            dydx = gt4.gradient(yz, obl1)
        dy2dx2 =  gt3.gradient(dydx, obl1)
    dy3dx3 = gt2.gradient(dy2dx2, obl1)
d4ydx4 = gt1.gradient(dy3dx3, obl2)

print('d4ydx4 =', d4ydx4)
```

**Задание 7**  
Решите задачу парной линейной регрессии при помощи модели TensorFlow, рассматривая тензор ранга 1 из пункта 1 задания как значения зависимой переменной (отклика), а точки отрезка из индивидуального задания как значения независимой переменной (предиктора). Предварительно масштабируйте независимую и зависимую переменные на интервал [0, 1]. Оцените качество полученной модели по показателю качества регрессии, указанному в индивидуальном задании. Количество эпох, скорость обучения и начальные значения весов выберите самостоятельно, обеспечивая сходимость обучения.

Максимальная ошибка (maximum residual error, MaxErr)

$$MaxErr=\max_{i=\overline{1,n}}\left|y_{i}-\hat{y}_{i}\right|$$


Масштабировала независимую и зависимую переменные на интервал [0, 1].

```python
def scale(x):
    return (x - tf.reduce_min(x)) / (tf.reduce_max(x) - tf.reduce_min(x))

xs = scale(rank_1_tensor)
ys = scale(y)
```

Определение модели, функции потерь и максимальной ошибки

```python
class Model(object):
    def __init__(self):
        self.w = tf.Variable(tf.constant(tf.random.normal(shape=())))
        self.b = tf.Variable(tf.constant(tf.random.normal(shape=())))

    def __call__(self, x):
        return self.w * x + self.b

model = Model()

def loss(predicted_y, target_y):
    return tf.reduce_mean(tf.square(predicted_y - target_y))

def maxerr(y_i, y_hat):
    errors = []
    n = len(y_i)
    for i in range(n):
        error = abs(y_i[i] - y_hat[i])
        errors.append(error)
    max_error = tf.reduce_max(errors)
    return max_error
```

Обучение модели

```python
TRUE_w = 0.1
TRUE_b = 0.3

# Функция для обучения модели
def train(model, inputs, outputs, learning_rate):
    with tf.GradientTape() as t:
        current_loss = loss(model(inputs), outputs)
    dw, db = t.gradient(current_loss, [model.w, model.b])
    model.w.assign_sub(learning_rate * dw)
    model.b.assign_sub(learning_rate * db)

    return current_loss

model = Model()

list_w, list_b = [], []
epochs = range(101)
losses = []
max_errors = []
for epoch in epochs:
    list_w.append(model.w.numpy())
    list_b.append(model.b.numpy())

    current_loss = train(model, xs, ys, learning_rate=0.1)
    losses.append(current_loss)

    predicted_xs = model(xs)
    max_error = maxerr(ys, predicted_xs).numpy()
    max_errors.append(max_error)

    print('Эпоха %4d: w=%1.2f b=%1.2f, потери=%2.5f, MaxErr ошибка: %2.5f' %
    (epoch, list_w[-1], list_b[-1], current_loss, max_error))
```

```python
print('\nMaxErr:', max_error)
```

```python
# Визуализация обучения
plt.plot(epochs, list_w, 'r', epochs, list_b, 'b')
plt.plot([TRUE_w] * len(epochs), 'r--', [TRUE_b] * len(epochs), 'b--')
plt.legend(['w', 'b', 'Истинное w', 'Истинное b'])
plt.show()
```
